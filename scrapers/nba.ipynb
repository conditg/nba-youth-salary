```python
from bs4 import BeautifulSoup, Comment
from urllib.request import urlopen
import numpy as np
import pandas as pd
from time import sleep
import csv
from string import ascii_lowercase
import random
```

```python
BASE_URL = "https://www.basketball-reference.com"


def get_player_links(section_url):
    html = urlopen(section_url).read()
    soup = BeautifulSoup(html, "html.parser")
    statst = soup.find(id="players")
    player_links = [BASE_URL + th.a["href"] for th in statst.tbody.find_all("th")]
    return player_links

```

```python
def get_stats(category_url):
    html = urlopen(category_url).read()
    soup = BeautifulSoup(html, "html.parser")
    saldiv = soup.find(id="all_all_salaries")
    salaries = []
    if saldiv is None:
        salaries = [{
            "playerid" : category_url.split("/")[5].split(".")[0],
            "season" : "no salary info",
            "team" : "no salary info",
            "salary" : 0
        }]
    else:
        saldiv2 = saldiv.contents[-2]
        soup2 = BeautifulSoup(saldiv2, "html.parser")
        saltable = soup2.find(id="all_salaries")
        for item in saltable.tbody.children:
            
            if len(item) <= 1:
                continue
            
            row = {
                "playerid" : category_url.split("/")[5].split(".")[0],
                "season" : item.find(attrs={"data-stat": "season"}).string,
                "team" : item.find(attrs={"data-stat" : "team_name"}).string,
                "salary" : item.find(attrs={"data-stat" : "salary"}).string
            }
            
            salaries.append(row)
            
        
    return salaries
```

```python
letter = ("https://www.basketball-reference.com/players/z/")
    
players = get_player_links(letter)
data = [] 
for player in players:
    sals = get_stats(player)
    data.append(sals)
    sleep(3) # be nice
```

```python
with open('zplayers.csv', 'w') as csvfile:
    fieldnames = ['playerid', 'season', 'team', 'salary']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for player in data:
        for season in player:
            writer.writerow(season)
```
